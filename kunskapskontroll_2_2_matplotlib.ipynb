### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# Figure: A container that holds all known elements that include axes, title etc. 
# Axe: A plot within. Contains data. Example: adding data, controlling axis scales and limits.
# Axis: Contains x, y, or z. 
# Artist is objects that can be drawn. Basically everything that is visible is an artist.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4)
y = x**2

#fig.ax = plt.subplot()
fig, ax = plt.subplots()
ax.set_title("A title, wow!")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.plot(x, y)
plt.show
```
